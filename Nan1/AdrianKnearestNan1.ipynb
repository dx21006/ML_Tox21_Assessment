```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
```

```python
# if y is a Series
y_binary_labelstrain = Nan0_ytrain_clean.iloc[:, 0].map({0: "non-toxic", 1: "toxic"})
y_binary_labelstest = Nan0_ytest_clean.iloc[:, 0].map({0: "non-toxic", 1: "toxic"})
```

```python
### Optimise metrices for pca and knn ###
from sklearn.model_selection import GridSearchCV

#Defining pipeline for optimisation
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("pca", PCA(n_components=0.95)),   # keep 95% variance
    ("knn", KNeighborsClassifier(n_neighbors=5))
])

scores = cross_val_score(pipeline, Nan0_xtrain_clean, Nan0_ytrain_clean, cv=5, scoring="balanced_accuracy")
print("Score for the model", scores.mean())

#PGrid Search optimisation
param_grid = {
    "pca__n_components": [5, 10, 20, 0.9, 0.95],
    "knn__n_neighbors": [3, 5, 7]
}

grid = GridSearchCV(pipeline, param_grid, cv=5, scoring="balanced_accuracy")
grid.fit(Nan0_xtrain_clean, Nan0_ytrain_clean)
print("Best params:", grid.best_params_)
print("Best CV accuracy:", grid.best_score_)
```

```python
# Best params: {'knn__n_neighbors': 7, 'pca__n_components': 0.95}
# Best CV accuracy: 0.7299971417435365

# Comparing with the original score for the model 0.7285089428781777
```

```python
#Now run the pipeline with the optimised model
pipeline_op = Pipeline([
    ("scaler", StandardScaler()),
    ("pca", PCA(n_components=0.95)),   # keep 95% variance
    ("knn", KNeighborsClassifier(n_neighbors=7))
])
pipeline_op.fit(Nan0_xtrain_clean, y_binary_labelstrain)
y_pred = pipeline_op.predict(Nan0_xtest_clean)

from sklearn.metrics import balanced_accuracy_score

print("Accuracy:", balanced_accuracy_score(y_binary_labelstest, y_pred))

print("\nConfusion Matrix:\n", confusion_matrix(y_binary_labelstest, y_pred))
```
