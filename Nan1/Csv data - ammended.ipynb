```python
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.ML.Descriptors import MoleculeDescriptors
```

```python
from rdkit.ML.Descriptors import MoleculeDescriptors
from rdkit.Chem import Descriptors
descriptor_names = [desc[0] for desc in Descriptors._descList]

# Create a calculator
calculator = MoleculeDescriptors.MolecularDescriptorCalculator(descriptor_names)
```

```python
def smiles_to_mol(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        return mol
    except:
        return None
######
def mol_to_descriptors(mol):
    try:
        return list(calculator.CalcDescriptors(mol))
    except:
        return [np.nan] * len(descriptor_names)
###Now Tandem them together###
def smiles_df_to_descriptors(df, smiles_col="smiles"):
    mols = df[smiles_col].apply(smiles_to_mol)
    
    desc_values = mols.apply(
        lambda mol: mol_to_descriptors(mol) if mol is not None 
        else [np.nan] * len(descriptor_names)
    )
    
    desc_df = pd.DataFrame(
        desc_values.tolist(),
        columns=descriptor_names,
        index=df.index
    )
    
    return desc_df
```

```python
X_trainNan0_Desc = smiles_df_to_descriptors(X_train0)
X_testNan0_Desc = smiles_df_to_descriptors(x_test0)
X_valNan0_Desc = smiles_df_to_descriptors(X_val0)

X_trainNan1_Desc = smiles_df_to_descriptors(X_train1)
X_testNan1_Desc = smiles_df_to_descriptors(x_test1)
X_valNan1_Desc = smiles_df_to_descriptors(X_val1)
```
